```python
# This exercise show For each product calculate the number of customers that bought it and their average age
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
products_df = pd.DataFrame({
 'Product_ID': [1, 2, 3],
 'Product_Name': ['iPhone 12', 'Samsung S21', 'Pixel 5'],
 'Product_Price': [999, 899, 699],
 })
display(products_df)
```

```python
sales_df = pd.DataFrame({
 'Sale_ID': [1, 2, 3, 4, 5, 6, 7],
 'Sale_Date': pd.to_datetime(['2021-01-01', '2021-02-01', '2021-02-15', '2020-03-01', '2020-04-01',
 '2020-02-15', '2020-01-01']),
 'Customer_ID': [1, 2, 3, 2, 3, 4, 4],
 'Product_ID': [1, 1, 2, 2, 3, 3, 1],
 })
display(sales_df)
```

```python
customers_df = pd.DataFrame({
 'Customer_ID': [1, 2, 3, 4],
 'Customer_Name': ['John Doe', 'Jane Doe', 'Jim Brown', 'Jake Smith'],
 'Customer_Age': [25, 30, 35, 20],
 'Customer_Email': ['john@doe.com', 'jane@doe.com', 'jim@brown.com', 'jake@smith.com'],
 })
display(customers_df)
```

```python
# Solution: 1
# Merge the sales and products dataframes
sales_customers = pd.merge(sales_df, customers_df, on='Customer_ID', how='inner')

# Merge the resualt with products dataframes
full_data = pd.merge(sales_customers, products_df, on='Product_ID', how='inner')
display(full_data)
```

```python
# Keep only the unique combination of Product-Customer, while keeping the age and product
unique_product_customers = full_data[['Product_ID', 'Product_Name', 'Customer_ID', 'Customer_Age']].drop_duplicates()
display(unique_product_customers)
```

```python
# Grouping by product and calculating aggregations
product_summary = unique_product_customers.groupby(['Product_ID', 'Product_Name']).agg(
    Unique_Customer_Count=('Customer_ID', 'count'), # Counting the IDs after filtering gives the number of unique customers
    Average_Customer_Age=('Customer_Age', 'mean')   # The average age of the unique customers
).reset_index()
display(product_summary)
```

```python
# Solution: 2
# calculate the total number of sales for each product per year
sales_df['Sale_Year'] = sales_df['Sale_Date'].dt.year # Extract Year, column must have a datetime-like data type (e.g., datetime64[ns])
sales_per_year = sales_df.groupby(['Product_ID', 'Sale_Year']).size().reset_index(name='Total_Sales')
display(sales_per_year)
```
