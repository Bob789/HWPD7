```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display

# Define data range
# x = np.linspace(0, 2 * np.pi, 10) # 100 points from 0 to 2*pi
x = np.arange(0, 5.1, 0.1) # 51 points from 0 to 5.0
#  Generate y-values
y1 = np.sin(x)
y2 = np.cos(x) 
y3 = np.sin(x + np.pi/4) # Phase shift by /4

print(len(x))  # Confirm number of points

# Optional: Check min/max
print("sin(x):", np.min(y1), "to", np.max(y1))  #  -1.0 to 1.0
print("cos(x):", np.min(y2), "to", np.max(y2))  #  -1.0 to 1.0

# Create the plot 
plt.figure(figsize=(10, 4)) # Optional: create figure with specific size

plt.subplot(1, 2, 1) # Create a 1x2 grid of subplots, and select the first subplot
#  Draw these x and y values on a chart plot sin(x).
plt.plot(x, y1, color='blue', linestyle='-', linewidth=1.5, label='sin(x)')
#  Draw these x and y values on a chart plot cos(x).
plt.plot(x, y2, color='red', linestyle='-.', linewidth=1.5, label='cos(x)')
# Add labels and title
plt.title('Simple Sine Wave using plt.plot()')
plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis (sin(x))')
plt.ylim(-1.1, 1.1) # Set y-axis limits
plt.xlim(0, 5) # Set x-axis limits
plt.grid(True) # Add grid lines
plt.legend()

# Second subplot: sin(x + /4)
plt.subplot(1, 2, 2)
plt.plot(x, y3, color='green', linestyle='--', linewidth=1.5, label='sin(x + /4)')
plt.title('Phase Shifted Wave')
plt.xlabel('X-axis (radians)')
plt.ylabel('Y-axis')
plt.ylim(-1.1, 1.1) # Set y-axis limits
plt.xlim(0, 5) # Set x-axis limits
plt.grid(True)
plt.legend()

# Show plot
plt.tight_layout() # Adjust layout to prevent overlap
#plt.subplots_adjust(wspace=0.5) # Adjust space between subplots
plt.show()
```
