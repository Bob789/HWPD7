```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
# Show in a new Dataframe how many licenses sold for each company from each product using pivoyt_table
# read file Sales_Funnel_CRM.csv
df = pd.read_csv('Sales_Funnel_CRM.csv')
display(df)
```

```python
# Show in a new Dataframe how many licenses sold for each company from each product using pivoyt_table
#    index='Company',      # Rows will be the unique company names
#    columns='Product',    # Columns will be the unique product names
#    values='Licenses',    # The values in the table cells will be from the 'Licenses' column
#    aggfunc='sum',        # Aggregate duplicate entries by summing the licenses
#    fill_value=0          # Replace any missing combinations (NaN) with 0
pivot_licenses = df.pivot_table(index='Company', columns='Product', values='Licenses', aggfunc='sum', fill_value=0).reset_index()
display(pivot_licenses )
```

```python
#  Show in a new Dataframe for each company what is the total amount of sale price and licences sold
pivot_sale_price_licences = df.pivot_table(index='Company', values=['Sale Price', 'Licenses'], aggfunc='sum').reset_index()
display(pivot_sale_price_licences)
```
